Μεθοδολογία Ερευνητικής Εργασίας "«Η βία κατά των μεταναστών και προσφύγων το 2015 και το 2020 από οργανωμένες ομάδες επιθέσεων της Χρυσής Αυγής»" 

1. Πρώτον, κατασταλάξαμε στο το θέμα της ερευνητικής εργασίας μας («Η βία κατά των μεταναστών και προσφύγων το 2015 και το 2020 από οργανωμένες ομάδες επιθέσεων της Χρυσής Αυγής»), καθώς, έμπνευση μας ήταν ρατσιστικές επιθέσεις στο περίγυρο μας.

2. Εύρεση πληροφοριών/δεδομένων και καταγραφή τους. Τα δεδομένα αντλήθηκαν από τις ετήσιες εκθέσεις του Δικτύου Καταγραφής Ρατσιστικής Βίας τις χρονολογίες 2015 (https://rvrn.org/wp-content/uploads/2022/05/Report_2015gr.pdf?fbclid=IwAR21dUF_hiw3aKTMobjqSPJlmC9wnZc0O86XeKd6_0j4v2DB8OnFv61BTVs) και 2020 (https://www.unhcr.org/gr/wp-content/uploads/sites/10/2021/05/GR_ETHSIA_EKTHESH.pdf?fbclid=IwAR2iZzzPY1ypmDKgz3xwzmuvRtUP6Hhok8wvfd_tNo6iDvB3rV0ABl3YkM4), ώστε να συλλέξουμε αναλυτικά τα καταγεγραμμένα περιστατικά βίας κατά των μεταναστών και προσφύγων από οργανωμένες ομάδες της Χρυσής Αυγής (Χ.Α.). 

3. Εκτενή μελέτη των δεδομένων και των πληροφοριών προκειμένου να επισημοποιηθεί ότι πρόκειται για έγκυρα δεδομένα και δεδομένα που είναι εφικτό να επεξεργαστούμε και να μελετήσουμε απο φορέα που εμπιστευόμαστε. 

4. Διατύπωση ερευνητικών ερωτημάτων και διερεύνηση πιθανών στοιχείών που μας λείπουν. 

5. Μελέτη μεθόδων συλλογής των δεδομένων. 

6. Απόπειρα περισυλλογής των δεδομένων προγραμματιστικά μέσω της εφαρμογής συλλογής δεδομένων από κειμενική μορφή Tika. 

7. Αλλαγή μεθόδου περισυλλογής δεδομένων δια χειρός με προσοχή μέσα από τις εκθέσεις και συγκέντρωση τους σε αρχείο google spreadsheet (excel). (Repository αρχείο: https://github.com/chrisoyla/PROJECT_CHRUSOYLA_ZEIMPEKAKI_KATERINA_DIMITRESKOY/blob/main/DATA_SET.csv)

8. Ομαδοποίηση στοιχείων σε κατηγορίες όπως: προέλευση θυμάτων 2015 και 2020, νομικό καθεστώς θυμάτων 2015 και 2020, φύλλο θυμάτων 2015 και 2020, εθνικότητες δραστών 2015 και 2020, φύλλο δραστών 2015 και 2020, ηλικία δραστών 2015 και 2020, ιδιότητα δραστών 2015 και 2020. Η διαδικασία αυτή βοήθησε στον καθαρισμό των δεδομένων ώστε να βρεθεί το συνολικό μέγεθος των μεταβλητών, να βρεθούν οι κενές τιμές και να γίνει μια πρώτη οπτικοποίηση των συνολικών δεδομένων. 

9. Δημιουργία αποθετηρίου και προσωπικών λογαριασμών στο Github. 

10. Δημιουργία πρώτου PITCH αρχείου που αναγράφει τις πρώτες σκέψεις για το θέμα. 

11. Συζήτηση θέματος μέσα στην ακαδημαϊκή τάξη: διορθώσεις, επανεξέταση θέματος, συμβουλές.

12. Μετά την ολοκήρωση της συλλογής των δεδομένων και της εκκαθάρισης τους, ξεκινά η δημιουργία του κώδικα και η οπτικοποίηση των δεδομένων. 

13. Επεξεργασία κώδικα και δημιουργία διαγραμμάτων για τα δεδομένα μέσω Jupyter Notebook. 

14. Αναζήτηση συνεντευξιαζόμενου για επιβεβαίωση ευρημάτων και απάντηση ερευνητικών ερωτημάτων και αποριών. 

15. Εύρεση συνεντευξιαζόμενου: Κώστας Παπαδάκης: δικηγόρος, συνήγορος πολιτικής αγωγής Χ.Α.

16. Δημιουργία στο αποθετήριο του GitHub: DATA_SET, DATA_BASE, READ_ME, DATA_VERSION_2

17. Συζήτηση μέσα στην ακαδημαϊκή τάξη: τελικές διορθώσεις στον κώδικα, συζήτηση των data και τελικές αποφάσεις. 

18. Διεξαγωγή συνέντευξης με τον συνεντευξιαζόμενο. 

19. Διόρθωση απαραιτήτων στον κώδικα και καθαρισμός δεδομένων. 

20. Συγκέντρωση διαθέσιμων πληροφοριών για τελικό story και ξεκαθάρισμα τους. 

21. Δημιουργία στο GitHub Repository: DATA_BASE_FINAL, DOCUMENTATION, SCRIPT, METHODOLOGIA

22. Δημιουργία τελικού story.
