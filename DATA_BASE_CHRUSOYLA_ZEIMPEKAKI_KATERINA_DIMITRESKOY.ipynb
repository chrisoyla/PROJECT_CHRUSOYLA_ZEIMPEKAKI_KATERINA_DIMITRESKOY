```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# region victim nationalities 2015

victim_nationalities_2015= [{'Egypt':7, 'Ethiopia':5, 'Afghanistan':11, 'Bulgaria':1, 'India':1, 'Iraq':1, 'Iran':3, 'Bangladesh':27, 'Nigeria':2, 'Pakistan':14, 'Sudan':1, 'Syria':3}]
print(victim_nationalities_2015)
```

```python
#region victim nationalities 2020

victim_nationalities_2020= [{'Afghanistan':10, 'Syria':9, 'Palestine':6, 'Sierra Leone':6, 'Mixed Nationalities':6, 'Egypt':2, 'Cameroon':2, 'Bangladesh':2, 'Côte d\'Ivoire':1, 'Guinea':1, 'Denmark':1, 'Iraq':1, 'Iran':1, 'Morocco':1, 'Pakistan':1, 'Syria/Palestine':1, 'Kurdish Origin':1}]
print(victim_nationalities_2020)
```

```python
# create data frames
victim_nationalities_2015_df = pd.DataFrame(victim_nationalities_2015)
victim_nationalities_2020_df = pd.DataFrame(victim_nationalities_2020)
victim_nationalities_df = pd.merge(victim_nationalities_2015_df, victim_nationalities_2020_df,
                                  how='outer', on='nationalities', left_on='nationalities_2015', right_on='nationalities_2020').fillna(0)
```

```python
# print table
print('Victim Nationalities')
print(victim_nationalities_df)
print('\n\n')
```

```python
# show plot
victim_nationalities_df.plot(
    title='Victim Nationalities')
plt.get_current_fig_manager().set_window_title('Victim Nationalities')
plt.show()
```
