Έμπνευση
Βάλτε εδώ ένα link προς ένα θέμα που έχετε δει να είναι ήδη δημοσιευμένο και λειτουργεί ως έμπνευση για εσάς για το συγκεκριμένο πρότζεκτ που θέλετε να κάνετε.
Βάλτε, επίσης, μια εικόνα ενός γραφήματος που έχετε δει, σας αρέσει και πιστεύετε ότι κάπως έτσι μπορείτε να παρουσιάσετε και εσείς την ανάλυσή σας. Το γράφημα αυτό μπορεί, φυσικά, να αφορά και διαφορετικό θέμα από το δικό σας.
Περαιτέρω πληροφορίες για το θέμα σας
Τίτλος εργασίας: γράψτε εδώ έναν πιθανό τίτλο για το θέμα σας

Δεδομένα: Εάν έχετε ήδη βρει δεδομένα για το θέμα σας, βάλτε εδώ τους υπερσυνδέσμους προς τα σχετικά dataset. Εάν πρόκειται να κάνετε scraping, εξηγήστε εδώ από πού θα συλλέξετε τα δεδομένα σας, παραπέμποντας στις σχετικές ιστοσελίδες

Πιθανά προβλήματα/ανασφάλειες/απορίες: Γράψτε εδώ πιθανά προβλήματα ή ανησυχίες που έχετε σχετικά με το θέμα σας
Ερευνητικά ερωτήματα
Σημειώστε εδώ σε λίστα τα ερευνητικά ερωτήματα που θέλετε να απαντήσετε με αυτό το πρότζεκτ.

Προέρευνα/δουλειά που έχετε ήδη κάνει στο πλαίσιο του σχεδιασμού του πρότζεκτ
Εξηγήστε συνοπτικά τη δουλειά που έχετε κάνει μέχρι τώρα, στο πλαίσιο της προετοιμασίας. Έχετε βρει δεδομένα; Τα έχετε κοιτάξει; Τα έχετε καθαρίσει;

Βάλτε εδώ μια εικόνα (π.χ. ένα γράφημα ή μια πρώιμη ανάλυση) που να δείχνει κάποιο πρώτο εύρημά σας, το οποίο σας κάνει να θέλετε να ερευνήσετε περαιτέρω/καλύτερα το θέμα
CHECKLIST
Σκεφτείτε εάν έχετε ολοκληρώσει τα βασικά βήματα για το pitch σας και βάλτε x στην παρακάτω λίστα για κάθε βήμα που έχετε ολοκληρώσει. Στο στάδιο της πρότασης θέματος που βρίσκεστε, όσα περισσότερα x έχετε, τόσο καλύτερα προτεοιμασμένοι είστε. Όσα λιγότερα x έχετε, τόσο περισσότερο πρέπει να σχεδιάσετε καλύτερα το πρότζεκτ σας.
