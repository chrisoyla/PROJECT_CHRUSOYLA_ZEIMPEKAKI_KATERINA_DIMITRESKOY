ΠΕΡΙΛΗΨΗ ΤΟΥ PROJECT

Το συγκεκρικέμνο Project επιδιώκει να συσχετίσει τα δεδομένα της έκθεσης του 2015 ( https://rvrn.org/wp-content/uploads/2022/05/Report_2015gr.pdf?fbclid=IwAR21dUF_hiw3aKTMobjqSPJlmC9wnZc0O86XeKd6_0j4v2DB8OnFv61BTVs) με τα δεδομένα της έκθεσης του 2020 (https://www.unhcr.org/gr/wp-content/uploads/sites/10/2021/05/GR_ETHSIA_EKTHESH.pdf?fbclid=IwAR2iZzzPY1ypmDKgz3xwzmuvRtUP6Hhok8wvfd_tNo6iDvB3rV0ABl3YkM4 ) έτσι ώστε να απαντήθει το βασικό ερευνητικό μας ερώτημα: Οι ρατσιστικές επιθέσεις έναντι των προσφύγων/μεταναστών έχουν μειωθεί ή αυξηθεί κατά τις χρονολογίες 2015 όπου η Χρυσή Αυγή συμμετείχε ακόμα στη Βουλή και 2020 με τη ρήξη του κόμματος (Εφετείο 2020); Σκοπός είναι η εξαγωγή των δεδομένων ή προγραμματιστικά μέσα απο το κείμενο ή γραπτά για τη δημιουργία ενός δικού μας data base και ύστερα η εκκαθάριση τους και η επεξεργασία τους έτσι ώστε να δημιουργηθεί κώδικας που να αναγράφει τα δεδομένα αυτά σε ραβδογραφήματα. Με αυτόν τον τρόπο τα γραφήματα αυτά θα εικονογραφήσουν τα δεδομένα για να γίνει ξεκάθαρη η απάντηση στο ερευνητικό μας ερώτημα και θα γίνει η τελική μορφή των αποτελεσμάτων. 

ΠΡΟΒΛΗΜΑΤΑ/ΑΛΛΑΓΕΣ

Στην αρχή στο pitch αναγραφόταν σαν θέμα οι ρατσιστικές επιθέσεις γενικά τις χρονολογίες του 2015 και του 2020 αλλά αργότερα άλλαξαν και συγκεκριμενοποιήθηκαν και έγιναν για τις ρατσιστικές επιθέσεις απέναντι σε μετανάστες/πρόσφυγες. 

Η προγραμματιστική εξαγωγή των δεδομένων από το κείμενο ήταν αρκετά απαιτητική και απαιτούσε χρόνο που δεν ήταν διαθέσιμος, μετά από αρκετές προσπάθειες η ομάδα αποφάσισε να διεξάγει τα δεδομένα γραπτά με μεγάλη προσοχή και απο τις δύο εκθέσεις, επίσης, κάποια δεδομένα μέσα απο τις εκθέσεις δεναναγράφονται έτσι ώστε να συμπληρώνουν τον συνολικό αριθμό δεδομένων απο κάθε κατηγορία, με αποτέλεσμα να μην μπορεί να γίνει σύγκριση όλων των δεδομένων μεταξύ τους. Έτσι, εκτός απο την εκτενή διερεύνηση των δεδομένων μέλημα μας είναι η απάντηση του ερωτήματος: Γιατί δεν αναγράφονται όλα τα δεδομένα στις εκθέσεις;
