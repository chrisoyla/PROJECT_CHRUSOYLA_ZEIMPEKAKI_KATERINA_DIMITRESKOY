        Η βία κατά των μεταναστών και προσφύγων από οργανωμένες ομάδες επιθέσεων της Χρυσής Αυγής (έτη 2015 και 2020).                        
	
	Στο πλαίσιο ακαδημαϊκής ποσοτικής και ποιοτικής έρευνας που διεξήχθη από τις τεταρτοετείς φοιτήτριες, Ζεϊμπεκάκη Χρυσούλα και Δημητρέσκου Κατερίνα του Παντείου Πανεπιστημίου, Κοινωνικών και Πολιτικών Επιστημών και τμήματος Επικοινωνίας, Μέσων και Πολιτισμού, προκύπτει το εξής ερώτημα προς διερεύνηση: Οι οργανωμένες ρατσιστικές επιθέσεις με θύματα μετανάστες και πρόσφυγες μειώθηκαν με την απόφαση του Εφετείου του 2020 και τη ρήξη του πολιτικού κόμματος Χρυσής Αυγής σε σύγκριση με το 2015, όπου το κόμμα συμμετείχε ακόμα στη Βουλή;
	Σύμφωνα με την ετήσια έκθεση του Δικτύου Καταγραφής Περιστατικών Ρατσιστικής Βίας (UNHCR) (https://www.unhcr.org/gr/wp-content/uploads/sites/10/2016/12/Report_2015gr.pdf) κατά τη περίοδο Ιανουαρίου-Δεκεμβρίου 2015, το Δίκτυο μέσω ανώνυμων μαρτυριών και συνεντεύξεων που καταγράφηκαν από τους φορείς του Δικτύου, κατέγραψε 79 περιστατικά ρατσιστικής βίας σε μετανάστες και πρόσφυγες λόγω εθνικής ή εθνοτικής προέλευσης, θρησκείας, ή χρώματος, ενώ, στα 4 από αυτά, αιτία αποτελούσε και ο σεξουαλικός προσανατολισμός. Η διαφορά της συγκεκριμένης έκθεσης με τις προηγούμενες (2011-2014) αποτέλεσε η χρηματοδότηση ειδικού προγράμματος που υλοποίησε το Ελληνικό Φόρουμ Μεταναστών αποκλειστικά για τις ρατσιστικές επιθέσεις. Η καταγραφή των περιστατικών αυτών θεωρείται η πιο απαιτητική σε σύγκριση με τις υπόλοιπες, τόσο λόγω των συνθηκών διαβίωσης των θυμάτων, αλλά και επειδή στις περιπτώσεις αυτές τα θύματα προτιμούν να μη μοιραστούν την εμπειρία τους, για την προστασία της ασφάλειας τους και της σωματικής τους ακεραιότητας. Αν και οι καταγραφές μοιάζουν αυξημένες από τα προηγούμενα έτη, λόγω των ενισχύσεων τους από φορείς που εστίασαν σε συγκεκριμένα περιστατικά, αποτελούν τη κορυφή του παγόβουνου. Σε σύγκριση με την Ετήσια Έκθεση του 2020 (https://www.unhcr.org/gr/wp-content/uploads/sites/10/2021/05/GR_ETHSIA_EKTHESH.pdf) , παρόλο που οι συνολικές καταγραφές των ρατσιστικών επιθέσεων μειώνονται απο 273 περιστατικά σε 107, οι καταγραφές των ρατσιστικών επιθέσεων ενάντια σε πρόσφυγες και μετανάστες παραμένουν στάσιμες σε σύνολο 78, εκ των οποίων οι 4 κατηγορήθηκαν και για τον σεξουαλικό τους προσανατολισμό. Η μεθοδολογία των καταγραφών παραμένει ίδια, αλλά, αλλάζει το κοινωνικό πλαίσιο, καθώς με επικράτηση της πανδημίας COVID-19 και τον περιορισμό των υπηρεσιών αλλά και των δομών των φορέων του Δικτύου σημειώνεται σημαντική υποαναφορικότητα.
	Για την εξαγωγή της έρευνας τα δεδομένα που συλλέχθηκαν από τις δύο εκθέσεις χωρίστηκαν σε κατηγορίες και συγκρίθηκαν μεταξύ τους:

    Αρχικά, εξετάστηκαν οι εθνικότητες των θυμάτων της έκθεσης του 2015 σε σύγκριση με την έκθεση του 2020. Το μεγαλύτερο αριθμό επιθέσεων δέχτηκαν τα θύματα με χώρα προέλευσης το Μπαγκλαντές με σύνολο 27 περιστατικά το 2015, ενώ το 2020 ο μεγαλύτερος αριθμός επιθέσεων προερχόταν από το Αφγανιστάν με 10 περιστατικά. Κοινές εθνικότητες στοχοποίησης αποτελούν και στις δύο εκθέσεις η Συρία (3 το ‘15, 9 το ‘20), το Πακισταν (14 το ‘15, 1 το ΄20), το Μπαγκλαντες (27 το ‘15. 2 το ΄20), το Ιράν (3 το ‘15, 1 το ΄20), το Αφγανιστάν (11 το ΄15, 10 το ΄20) και η Αίγυπτος (7 το ΄15, 2 το ΄20), ενώ,  για 3/79 της έκθεσης του ‘15 δεν γνωρίζουμε την εθνική προέλευση τους όπως και για 26/78 της έκθεσης του ‘20:
    

    Ύστερα, το νομικό καθεστώς των θυμάτων της έκθεσης του 2015 σε σύγκριση με την έκθεση του 2020. Το έτος 2015 η πλειοψηφία των κατεγραμμένων θυμάτων (σύνολο 46) δεν κατέχουν κάποιο νομικό καθεστώς, ενώ το 2020 η πλειοψηφία ζητούν άσυλο (σύνολο 28), ενώ, 4/79 από τα καταγεγραμμένα θύματα της έκθεσης του ‘15 δεν αναγράφεται το νομικό καθεστώς που εντάσσονται όπως και 26/78 της έκθεσης του ‘20:


Τρίτη μεταβλητή αποτέλεσε το φύλλο των θυμάτων της έκθεσης του 2015 σε σύγκριση με την έκθεση του 2020. Το έτος 2015 τα θύματα αποτελούν κατά κύριο λόγω άντρες (σύνολο 70) όπως και το έτος 2020 (σύνολο 40), ενώ, αξιοσημείωτη είναι η ύπαρξη μεικτών ομάδων (σύνολο 37):


Όσον αφορά τους δράστες, μελετήθηκαν και οι δικές τους εθνικότητες από την  έκθεση του 2015 σε σύγκριση με την έκθεση του 2020. Το ‘15 η πλειοψηφία ήταν Έλληνες (σύνολο 71) όπως και το ‘20 (σύνολο 57). Για 5/79 από τους δράστες της έκθεσης του ΄15 δεν γνωρίζουμε την εθνικότητα τους όπως και για 4/78 από την έκθεση του ΄20:


Μελετήθηκε και το φύλλο των δραστών όπως αναφέρονται στην έκθεση του 2015 σε σύγκριση με την έκθεση του 2020. Και στις δύο εκθέσεις η πλειοψηφία των δραστών είναι άντρες (66 το ‘15 και 57 το ‘20), ενώ, δεν γνωρίζουμε το φύλλο για 7/79 δράστες το 2015 και για 4/78 δράστες το 2020:


Αλλά, μελετήθηκαν και οι ηλικίες των δραστών συγκριτικά με τις εκθέσεις των δύο χρονολογιών. Η μεγαλύτερη ηλικιακή ομάδα δραστών ήταν 30-40 το 2015 (σύνολο 24 περιστατικά) και μέχρι και 75 έτη το 2020 (σύνολο 75 περιστατικά). Ενώ, δεν γνωρίζουμε την ηλικιακή ομάδα για 12 από τους δράστες το 2015.


Και τέλος, οι ιδιότητες των δραστών στην πλειοψηφία τους ήταν απλοί πολίτες το 2015 (σύνολο 49) και ντόπιοι-οργανωμένες ομάδες πολιτών το 2020 (σύνολο 75), ενώ, δεν γνωρίζουμε τις ιδιότητες των δραστών από 6 περιστατικά το 2015. 


    Ωστόσο, όπως μας ενημέρωσε, μέσα από την συνέντευξη που διεξήχθη, ο κύριος Κ. Παπαδάκης (δικηγόρος, συνήγορος πολιτικής αγωγής της Χρυσής Αυγής) δεν μπορούμε να έχουμε μία ολοκληρωμένη εικόνα για τον ακριβή αριθμό των περιστατικών. Αυτό οφείλεται, διότι πολλές φορές η ίδια η Αστυνομία προκαλούσε κωλύματα ως προς την αντιμετώπιση των επιθέσεων, “αντιμετώπιζαν κατά κύριο λόγο μια αστυνομία εχθρική προς αυτούς και φιλική προς τους δράστες π.χ. τα Α.Τ. Παντελεήμονα, Νίκαιας, Ομόνοιας κ.α., τα οποία ενεργούσαν πολλές φορές συγκαλυπτικά.” Επιπλέον, τα θύματα των ρατσιστικών επιθέσεων πολλές φορές δεν γνωστοποιούσαν τις επιθέσεις που γινόντουσαν εις βάρος τους, αφού ενισχυόνταν το αίσθημα του φόβου. Υπήρχαν φορές δηλαδή που κρατούνταν στο τμήμα (κυρίως όσοι δεν είχαν λύσει το πρόβλημα μονιμοποίησης στην Ελλάδα) και απειλούνταν απο τις αρχές με απέλαση. Ένα άλλο πρόβλημα που αντιμετώπιζαν απο τις ίδιες τις αρχές, είναι ότι πολλοί αστυνομικοί, όχι μόνο υποστήριζαν την Χρυσή Αυγή, αλλά συμμετείχαν και στις δράσεις που διοργάνωνε, “Υπάρχουν αστυνομικοί που συμμετείχαν χωρίς τη στολή τους σε εκδηλώσεις της ΧΑ και η ΧΑ γενικώς αισθανόταν ως μία αστυνομία χωρίς στολή, η οποία μπορούσε να πάρει το νόμο στα χέρια της και να κάνει αυτό που η ίδια θεωρούσε δίκαιο, νόμιμο, χωρίς να τηρεί τους νόμους και τα δικαιώματα των ανθρώπων.”Φυσικά, όλες αυτές οι πρακτικές, δεν έρχονται σε αντίθεση ως προς αυτά που πρέσβευε το κόμμα τους, «Εξισώνετε τους ομογενείς μετανάστες με τον κάθε λογής υπάνθρωπο που έχει εισβάλλει στην πατρίδα μας, με τις κάθε λογής αρρώστιες που κουβαλάει», Ελ. Ζαρούλια, βουλευτής της ΧΑ, σε ομιλία στην Βουλή. 
    Επιπλέον, οι ρατσιστικές απόψεις εξακολουθούσαν να ριζώνουν στις συνειδήσεις τους, κάτι που προέρχεται από την γενικότερη Ιδεολογία του φασιστικού κόμματος, “Οι σημερινοί κάτοικοι της χώρας, οι αληθινοί και γνήσιοι Έλληνες, αποτελούν την αυθεντική συνέχεια των Αρχαίων Ελλήνων όπως έχει αποδείξει σήμερα πλέον εκτός της Ιστορίας, της γλωσσολογίας, της λαογραφίας και η γενετική επιστήμη και η ανθρωπολογία. Αυτό σε πείσμα των διεθνιστών, φιλελεύθερης και μαρξιστικής κατευθύνσεως, οι οποίοι με κάθε τρόπο αμφισβητούν αυτή την Αλήθεια! Την αμφισβητούν όχι μόνο στα λόγια, αλλά και στην πράξη, με τα εκατομμύρια των λαθρομεταναστών, στους οποίους επιθυμούν να δώσουν και «ελληνική ιθαγένεια», να τους «ελληνοποιήσουν» και να σπάσουν έτσι την φυλετική συνέχεια του Έθνους μας, μια συνέχεια τόσων χιλιάδων ετών.» απόσπασμα από την επίσημη ιστοσελίδα της Χ.Α., Ιδεολογία.
    Απ’ αυτό, διαφαίνεται καθαρά το μίσος τους απέναντι στους ξένους. Επομένως, η πεποίθηση της καθαρότητας του έθνους, έρχεται να συναντήσει τα χαρακτηριστικά συνθήματα των υποστηρικτών της ΧΑ “Έξω οι ξένοι”. 
    Για αυτούς, ο εχθρός είναι κοινός και είναι ο ξένος. Δεν υπάρχει διαχωρισμός στους ξένους. Δεν υπάρχουν δηλαδή μετανάστες, πρόσφυγες, ή αιτούντες άσυλο. Γι’ αυτούς, είναι ο ξένος εχθρός, είναι οι  λαθρομετανάστες που προσπαθούν να βρωμίσουν τη χώρα τους, την οποία και έπρεπε να καθαρίσουν, «…η κακοποίηση των αλλοδαπών Πακιστανών από εμάς, ήταν διασκέδαση», κατάθεση πρώην μέλους ΧΑ.
    Όλη αυτή η βία που κρύβεται πίσω απ’ το αίσθημα δικαίου σε συνάρτηση με τον ρατσισμό, κάτι που ο κύριος Κ. Παπαδάκης εξηγεί αναλυτικά, “Υπάρχει μία μερίδα εκπήγασης ας πούμε, ένας αυθόρμητος κοινωνικός ρατσισμός, που υπάρχει στη συνείδηση κάθε ανθρώπου και ο οποίος εκφράζεται ανεξαρτήτως πολιτικής συγκυρίας. Υπάρχει όμως και ένας οργανωμένος πολιτικός ρατσισμός. Μία οργανωμένη ρατσιστική ιδεολογία, η οποία όταν έχει την αίσθηση ότι έχει ανταπόκριση στο κοινωνικό σώμα, δημιουργεί δυναμική. Τι σημαίνει αυτό; Εγώ πχ δεν είμαι χρυσαυγίτης, άλλα όταν βλέπω τα τάγματα εφόδου να χτυπάνε και τα λοιπά, ε πάω κι εγώ, τσαμπουκαλεύομαι στον διπλανό μου που είναι μετανάστης, του ρίχνω μία σφαλιάρα και νιώθω ότι κάτι είμαι. Κι αυτό είναι μια ρατσιστική επίθεση. Μπορεί βέβαια να μη χρεώνεται άμεσα στη ΧΑ, αλλά εγώ την χρεώνω έμμεσα. Διότι αυτή, δημιουργεί πιάτσα, δημιουργεί συνθήκες ευρύτερης συμπεριφοράς ανοχής και νομιμοποίησης ρατσιστικών επιθέσεων.“


(Κ. Παπαδάκης κατά τη διάρκεια της καταδίκης της Χ.Α., Φωτογραφία από το site “Φοιτητικός Κόσμος”) 


    Σ’ αυτό το σημείο, αξίζει να σημειωθεί πως στις 2 εκθέσεις, τα περιστατικά βίας κατά μεταναστών και προσφύγων φαίνεται από το 2015 έως και το 2020 να μειώνονται κατά ένα. Παρόλα αυτά, εκτός του ότι δεν ανταποκρίνεται ο αριθμός των περιστατικών σε σχέση με τις καταγραφές εξαιτίας του φόβου των θυμάτων αλλά και της στάσης ορισμένων αρχών όπως αναγράφεται παραπάνω, ωστόσο ο κύριος Παπαδάκης διαβεβαιώνει για την σημαντική μείωση επιθέσεων από την καταδίκη της Χ.Α. και μετά. Αν αναλογιστεί κανείς ότι το 2015 συναντούσαμε τα τάγματα εφόδου που απαρτίζονταν απο 10 έως και 40 άτομα (“Διαμόρφωσε σε κάποιες συνοικίες, ομάδες εκπαιδευμένες που λειτουργούσαν σαν στρατιωτικοί λόχοι, με στρατιωτικές στολές, ρόπαλα, κράνη, κάνανε εκπαίδευση στις πολεμικές τέχνες και εμφανιζόντουσαν σε διάφορες περιοχές είτε για να κάνουν επίδειξη δύναμης, είτε για να κάνουν συσσίτια μόνο για Έλληνες και ταυτόχρονα παίρνανε ταυτότητες για να κάνουν στρατολόγηση και να δείχνουν τη δύναμή τους, είτε για να προκαλούν μετανάστες και όχι μόνο”) με εντολή του κόμματος, ενώ από το 2020 και μετά, οι επιθέσεις μειώθηκαν και πλέον γίνονται είτε ατομικά, είτε από ομάδες 3-4 ατόμων. 


(Επίθεση τάγματος εφόδου από Χρυσαυγίτες σε μετανάστες, Φωτογραφία από άρθρο ΕΦ.ΣΥΝ.)

    Τέλος, όταν ρωτήσαμε τον κύριο Παπαδάκη γιατί τα ποσοστά μείωσης δεν είναι αρκετά ικανοποιητικά μας απάντησε πως μία δικαστική απόφαση όπως η καταδίκη της Χ.Α. είναι ικανοποιητική αλλά όχι αρκετή για να εξαλείψει το φαινόμενο της ρατσιστικής βίας. Πόσο μάλλον όταν μια μεγάλη μερίδα των ΜΜΕ συνεχίζει την προπαγάνδα και την συνωμοσιολογία κατά των μεταναστών και προσφύγων, υπάρχει ο κίνδυνος επανεμφάνισης του φαινομένου.
